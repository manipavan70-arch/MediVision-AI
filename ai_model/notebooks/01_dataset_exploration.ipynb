```python
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
dataset_path = r"C:\Users\manip\OneDrive\Desktop\MediVision-AI\dataset"
```

```python
train_path = os.path.join(dataset_path, "train")

fracture_path = os.path.join(train_path, "fracture")
normal_path = os.path.join(train_path, "normal")

print("Train Path:", train_path)
print("Fracture Folder:", fracture_path)
print("Normal Folder:", normal_path)
```

```python
fracture_images = os.listdir(fracture_path)
normal_images = os.listdir(normal_path)

print("Number of Fracture Images :", len(fracture_images))
print("Number of Normal Images :", len(normal_images))
```

```python
image = Image.open(os.path.join(fracture_path, fracture_images[0]))

plt.figure(figsize=(6,6))
plt.imshow(image)
plt.title("Fracture X-ray")
plt.axis("off")
plt.show()
```

```python
image = Image.open(os.path.join(normal_path, normal_images[0]))

plt.figure(figsize=(6,6))
plt.imshow(image)
plt.title("Normal X-ray")
plt.axis("off")
plt.show()
```

```python
print("Image Size :", image.size)
print("Image Mode :", image.mode)
```

```python
image = image.resize((224, 224))

plt.figure(figsize=(5,5))
plt.imshow(image)
plt.title("Resized Image (224 x 224)")
plt.axis("off")
plt.show()
```

```python
import numpy as np

image_array = np.array(image)

print("Shape:", image_array.shape)
print("Data Type:", image_array.dtype)
```

```python
normalized_image = image_array / 255.0

print("Minimum Pixel Value:", normalized_image.min())
print("Maximum Pixel Value:", normalized_image.max())
```

```python
import torch
from torchvision import transforms

transform = transforms.ToTensor()

tensor_image = transform(image)

print("Tensor Shape:", tensor_image.shape)
```
